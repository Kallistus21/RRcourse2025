```python
import os
import pandas as pd

#we set the path and specify the "data" so it's dynamic
os.chdir("C:\\Users\\Piotr\\Desktop\\RRcourse2025")
data_path = "Data"

#load the onet tasks csv data
task_data = pd.read_csv(f"{data_path}\\onet_tasks.csv")

#a function to load all the ISCO sheets into a list of DataFrames
def load_isco_data(excel_file, sheet_prefix="ISCO", num_sheets=9):
    return [pd.read_excel(excel_file, sheet_name=f"{sheet_prefix}{i}") for i in range(1, num_sheets + 1)]

#load the data
isco_file = f"{data_path}\\Eurostat_employment_isco.xlsx"
isco_dataframes = load_isco_data(isco_file)

#a unction to calculate total employment for a given country
def calculate_total_employment(country, isco_dfs):
    return sum(df[country] for df in isco_dfs)
#dynamic list of countries to process
countries = ["Belgium", "Spain", "Poland"]
#calculate employment per country
employment_totals = {country: calculate_total_employment(country, isco_dataframes) for country in countries}
#individual totals
total_Belgium = employment_totals["Belgium"]
total_Spain = employment_totals["Spain"]
total_Poland = employment_totals["Poland"]
```

```python
import pandas as pd

#assign ISCO levels with a loop
for i, df in enumerate(isco_dataframes, start=1):
    df['ISCO'] = i

#combine the datasets
all_data = pd.concat(isco_dataframes, ignore_index=True)

# add employment per country
for country in countries:
    all_data[f"total_{country}"] = pd.concat([employment_totals[country]] * len(isco_dataframes), ignore_index=True)

# and now calculate employment share PER country
for country in countries:
    all_data[f"share_{country}"] = all_data[country] / all_data[f"total_{country}"]
```

```python
import pandas as pd
import numpy as np

#we will extract the first digit
task_data["isco08_1dig"] = task_data["isco08"].astype(str).str[:1].astype(int)

#calculate the mean
aggdata = task_data.groupby("isco08_1dig").mean().drop(columns=["isco08"])
#construct a list of non-routine cognitive analytical tasks to track
non_routine_tasks = [
    "t_4A2a4",  
    "t_4A2b2",  
    "t_4A4a1"   
]

#merge data
combined = pd.merge(all_data, aggdata, left_on="ISCO", right_on="isco08_1dig", how="left")

#a function to standardize by country
def standardize_by_country(df, countries, task_columns):
    for country in countries:
        for col in task_columns:
            # Standardize: mean 0, std dev 1
            df[f"standardized_{col}_{country}"] = (df[col] - df[f"share_{country}"].mean()) / df[f"share_{country}"].std()
    return df

#and standardize columns by country
combined = standardize_by_country(combined, countries, non_routine_tasks)
```

```python
from scipy.stats import gmean

#list of tasks and columns
tasks = ["t_4A2a4", "t_4A2b2", "t_4A4a1"]
countries = ["Belgium", "Poland", "Spain"]

#a function to quickly calculate standardized values
def standardize_task(task, countries):
    for country in countries:
        #weighted mean and std dev
        temp_mean = np.average(combined[task], weights=combined[f"share_{country}"])
        temp_sd = np.sqrt(np.average((combined[task] - temp_mean) ** 2, weights=combined[f"share_{country}"]))
        
        #create columns
        combined[f"std_{country}_{task}"] = (combined[task] - temp_mean) / temp_sd

#standardize tasks
for task in tasks:
    standardize_task(task, countries)
```

```python
from scipy import stats
import matplotlib.pyplot as plt

#list of countries and tasks
countries = ["Belgium", "Poland", "Spain"]
tasks = ["t_4A2a4", "t_4A2b2", "t_4A4a1"]

#calculate NRCA for each country 
for country in countries:
    combined[f"{country}_NRCA"] = sum(combined[f"std_{country}_{task}"] for task in tasks)

#standardize NRCAs
def standardize_NRCA(country):
    temp_mean = np.average(combined[f"{country}_NRCA"], weights=combined[f"share_{country}"])
    temp_sd = np.sqrt(np.average((combined[f"{country}_NRCA"] - temp_mean)**2, weights=combined[f"share_{country}"]))
    combined[f"std_{country}_NRCA"] = (combined[f"{country}_NRCA"] - temp_mean) / temp_sd

for country in countries:
    standardize_NRCA(country)

#weighted mean for each country
for country in countries:
    combined[f"multip_{country}_NRCA"] = combined[f"std_{country}_NRCA"] * combined[f"share_{country}"]

#group by time and sum of weighted vlaues
agg_data = {}
for country in countries:
    agg_data[country] = combined.groupby(["TIME"])[f"multip_{country}_NRCA"].sum().reset_index()

#now plot the results
for country in countries:
    plt.plot(agg_data[country]["TIME"], agg_data[country][f"multip_{country}_NRCA"])
    plt.xticks(range(0, len(agg_data[country]), 3), agg_data[country]["TIME"][::3])
    plt.title(f"NRCA Intensity over Time: {country}")
    plt.show()
```
